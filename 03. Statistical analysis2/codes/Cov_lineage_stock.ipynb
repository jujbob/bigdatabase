## 1. 각 샘플마다 search에 대한 편차구하기

```python
df["dev_search"] = df["num_search"] - df["num_search"].mean()
```

## 2. 각 샘플마다 price에 대한 편차구하기

```python
df["dev_price"] = df["stock_price"] - df["stock_price"].mean()
```

## 3. 각 샘플마다 search값의 편차를 표준편차로 나누기

```python
df["covar_search"] = df["dev_search"] / np.std(df["num_search"])
```

## 4. 각 샘플마다 price값의 분산을 표준편차로 나누기

```python
df["covar_price"] = df["dev_price"] / np.std(df["stock_price"])
```

```python
df["covar_total"] = df["covar_search"] * df["covar_price"]
```
