```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x_train = np.array([[2, 3, 1], 
                    [4, 8, 9], 
                    [6, 2, 7], 
                    [8, 4, 6]])
y_train = np.array([[3], 
                    [7], 
                    [5], 
                    [8]])

n_samples = x_train.shape[0]
n_features = x_train.shape[1]
losses = []
```

```python
import torch
```

```python
a = torch.Tensor(3,1)
a.requires_grad
#a.size()
```

```python
weight = torch.nn.Parameter(torch.Tensor(3,1))
bias = torch.nn.Parameter(torch.Tensor(1))
print(weight)
print(weight.size())
#linear_model = weight
#print(linear_model)
#print(linear_model.parameters())
#weight
```

```python
x_train_torch = torch.from_numpy(x_train).float()
y_train_torch = torch.from_numpy(y_train).float()
print(x_train_torch)
print(x_train_torch.shape)
print(x_train_torch.requires_grad)
```

```python
optimizer = torch.optim.SGD([weight, bias], lr=0.001)
loss_function = torch.nn.MSELoss()
```

```python
for idx in range(100):
  optimizer.zero_grad()
  predicted_y = x_train_torch.mm(weight) + bias
  loss = loss_function(predicted_y, y_train_torch)  
  print(loss.item())
  loss.backward()
  optimizer.step()
```
