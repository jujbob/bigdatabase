# Multi-variable Linear Regression 수능 점수대비 등급 예측
 - input: 수학, 영어, 과학 공부시간 [2,3,1]
 - output: 수능점수 [3]
 - model: LR

## 1. 패키지 import

```python
import pandas as pd
import numpy as np
#from matpl
```

## 2. data preprocessing

```python
x_train = np.array([[2,3,1],
                   [4,8,9],
                   [6,2,7],
                   [8,4,6]])
y_train = np.array([
                    [3],
                    [7],
                    [5],
                    [8]
                    ])
```

```python
n_samples = x_train.shape[0]
n_features = x_train.shape[1]
```

## 3. Build Model (Hyo Function) 

```python
W = np.ones((3,1))
b = np.ones((1,1))
```

```python
def f(X):
  return np.matmul(x_train,W) + b
```

## 4. Cost function 설계

```python
def loss_function(y_pred, y_train):
  loss = 0
  #MSE
  dev = y_pred-y_train
  dev_sq = dev**2
  loss = dev_sq.sum() / len(y_pred)
  return loss
```

```python
y_pred = f(x_train)
```

```python
loss_function(y_pred, y_train)
```

```python
def gradient(x_train, y_train, y_pred):
  dw = 0
  db = 0
  n_samples = x_train.shape[0]
  dw = (1/n_samples) * np.matmul( x_train.transpose(), (y_pred - y_train) )
  db = (1/n_samples) * np.sum(y_pred - y_train, axis=0, keepdims=True)
  #print("dw", dw)

  return dw, db
```

```python
dw, db = gradient(x_train, y_train, y_pred)
```

```python
def weight_update(W, b, learning_rate, dw, db):
  W = W - learning_rate*dw
  b = b - learning_rate*db

  return W, b
```

```python
W, b = weight_update(W, b, 0.1, dw, db)
```

```python
def train():
  for idx in range(100):
    y_pred = f(x_train)
    loss = loss_function(y_pred, y_train)
    dw, db = gradient(x_train, y_train, y_pred)
    W, b = weight_update(W, b, 0.001, dw, db)

    print("loss값은:", loss, W, b)
```
