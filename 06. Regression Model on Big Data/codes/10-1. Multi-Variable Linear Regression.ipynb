## 3. 선형회귀의 평가: Mean Squred Error (MSE)
image.png

````python
def loss_function(y_pred, y_train):
  n_samples = y_train.shape[0]
  deviation = y_pred - y_train
  deviation_squred = deviation**2
  #print(deviation_squred)
  loss = (1/(2*n_samples)) * np.sum(deviation_squred)
  return loss
````

## 4. 선형회귀 오류의 피드백: Gradient Descent Algorithm
image.png

````python
def gradient(x_train, y_train, y_pred):
  n_samples = x_train.shape[0]
  dw = (1/n_samples) * np.matmul( x_train.transpose(), (y_pred - y_train) )
  db = (1/n_samples) * np.sum(y_pred - y_train, axis=0, keepdims=True)
  print("dw", dw)

  return dw, db

def gradient_descent(W, b, learning_rate, dw, db):
  W = W - learning_rate*dw
  b = b - learning_rate*db

  return W, b
````

## 5. Training with Multi-Variable의 연산


image.png

````python
for num_iter in range(100):

  print("x_train: ", x_train, x_train.shape)
  print(type(x_train))
  y_pred = f(x_train)
  print("y_pred:", y_pred, y_pred.shape)
  loss = loss_function(y_pred, y_train)
  losses.append(loss)
  dw, db = gradient(x_train, y_train, y_pred)
  W, b = gradient_descent(W, b, 1e-3, dw, db)
  print(W)
  print(loss)

# plot the cost
costs = np.squeeze(losses)    
plt.plot(losses)
plt.ylabel('Mean Squred Error')
plt.xlabel('The number of iteration')
plt.title("Multi-variable Linear Regression")
plt.show()
````
